# created Function for cleaning data

```python
def Rnewline(lst):
    new=[]
    for i in lst:
        new.append(i.strip().replace("\n",""))
    return new
```

# Get all the Audio book Name 

```python
name=soup.find_all("h3",class_="bc-heading bc-color-link bc-pub-break-word bc-size-medium")
name
```

```python
name1=[]
for i in range(0,len(name)):
    name1.append((name[i].get_text()).replace(" ",""))

AU_name=Rnewline(name1)
AU_name
```

# Get the writer of the book

```python
wrt=soup.find_all("li",class_="bc-list-item authorLabel")
wrt
```

```python
wrt1=[]
for i in range(0,len(wrt)):
    wrt1.append((wrt[i].get_text()).replace("Written by:\n", ""))
    
AU_wrt=Rnewline(wrt1)
AU_wrt
len(AU_wrt)
```

# Get the Audio book length

```python
time=soup.find_all("li",class_="bc-list-item runtimeLabel")
time
```

```python
time1=[]
for i in range(0,len(time)):
    time1.append((time[i].get_text()).replace("Length:",""))
time1
```

```python
AU_time= Rnewline(time1)
```

```python
AU_time
```

# Get the Audio book Language

```python
lan=soup.find_all("li",class_="bc-list-item languageLabel")
lan
```

```python
lan1=[]
for i in range(0,len(lan)):
    lan1.append((lan[i].get_text()).replace("Language:",""))

len(lan1)
```

```python
AU_lan= Rnewline(lan1)
AU_lan
```

# Get the Audio book Release date

```python
rdate=soup.find_all("li",class_="bc-list-item releaseDateLabel")
rdate
```

```python
rdate1=[]
for i in range(0, len(rdate)):
    rdate1.append((rdate[i].get_text()).replace("Release Date:",""))
    
```

```python
AU_Rdate=Rnewline(rdate1)
AU_Rdate
```

```python
rate=soup.find_all("li",class_="bc-list-item ratingsLabel")
rate
```

```python
rate1=[]
for i in range(0,len(rate)):
    rate1.append(rate[i].get_text())
    
AU_rating=Rnewline(rate1)
AU_rating
```

# Get the Price of the book

```python
price=soup.find_all("span",class_="bc-text bc-size-base bc-color-base")
price
```

```python
price1=[]
for i in range(0,len(price)):
    price1.append((price[i].get_text()).replace("Sale price:","").replace("Regular price:","").replace("or 1 credit",""))

r=Rnewline(price1)
r
#len(r)
```

```python
AU_price=[]
for i in range(1,60,3):
    AU_price.append(r[i])
    
AU_price
```
